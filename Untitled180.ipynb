Renaming columns in the merged table as the following:

# Finding cumulative reaction scores

Initialize 16 variables and increment their scores (content categories)

```python
contents_categories = [c.capitalize() for c in contents_categories]
contents_categories = list(set(contents_categories))
contents_categories
```

```python
scores = dict.fromkeys(contents_categories,0)
print(scores)
```

```python
merged_contents_reaction['Category'] = merged_contents_reaction['Category'].str.replace('"', '').str.capitalize()
```

```python
merged_contents_reaction['Category'].unique()
```

```python
def calc_scores(k: pd.DataFrame, scores: dict):
    for index, row in k.iterrows():
        if row['Reaction Type'] == 'heart':
            scores[row['Category']] += 60
        elif row['Reaction Type'] == 'want':
            scores[row['Category']] += 70
        elif row['Reaction Type'] =='disgust':
          scores[row['Category']] += 0
        elif row['Reaction Type'] =='hate':
          scores[row['Category']] += 5
        elif row['Reaction Type'] == 'interested':
          scores[row['Category']] += 30
        elif row['Reaction Type'] == 'indifferent':
          scores[row['Category']] += 20
        elif row['Reaction Type'] == 'love':
          scores[row['Category']] +=65
        elif row['Reaction Type'] == 'super love':
          scores[row['Category']] += 75
        elif row['Reaction Type'] == 'cherish':
          scores[row['Category']] += 70
        elif row['Reaction Type'] == 'adore':
          scores[row['Category']] += 72
        elif row['Reaction Type'] == 'like':
          scores[row['Category']] += 50
        elif row['Reaction Type'] == 'dislike':
          scores[row['Category']] += 10
        elif row['Reaction Type'] == 'intrigued':
          scores[row['Category']] += 45
        elif row['Reaction Type'] == 'peeking':
          scores[row['Category']] += 35
        elif row['Reaction Type'] == 'scared':
          scores[row['Category']] += 15
        else:
          scores[row['Category']] += 12
    return k

p = calc_scores(merged_contents_reaction,scores)
p = pd.DataFrame(p)

print(scores)
```
